```python
#a
import numpy as np
import matplotlib.pyplot as plt
import math
F0 = 500
Fs = 5000
N = 100
n = np.arange(N)
xn = np.sin(2 * math.pi * F0 * n / Fs)
plt.stem(n, xn)
plt.xlabel("n")
plt.show()
```

```python
#b2
import numpy as np
import matplotlib.pyplot as plt
import math
F0 = 2000
Fs = 50000
N = 26
n = np.arange(N)
filter_arr = []
for element in n:
    if element % 2 == 0:
        filter_arr.append(True)
    else:
        filter_arr.append(False)
n = n[filter_arr]
print(n)
xn = np.sin(2 * math.pi * F0 * n / Fs)
plt.stem(n, xn)
plt.show()
```
